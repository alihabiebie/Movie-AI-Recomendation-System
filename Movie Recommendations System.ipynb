```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('movie_dataset.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
fitur = ["genres", "keywords", "cast", "director"]
```

```python
def combo_feature(x):
    return x['genres'] + " " + x['keywords'] + " " + x['cast'] + " " + x['director']
```

```python
for i in fitur:
    df[i] = df[i].fillna("")
```

```python
df['combo_feature'] = df.apply(combo_feature, axis=1)
```

```python
df.head(11)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
cv_matrix = cv.fit_transform(df['combo_feature'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos_score = cosine_similarity(cv_matrix)
```

```python
def title_by_index(ind):
    return list(df[df.index == ind]['title'])

def index_by_title(title):
    return list(df[df.title == title]['index'])[0]
```

```python
similar_movie = list(enumerate(cos_score[movie_index]))
```
